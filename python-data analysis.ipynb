```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 
%matplotlib inline
df= pd.read_csv('/Users/jamesrich/Desktop/movies.csv')
df.head()

```

```python
for col in df.columns:
    percent_missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,percent_missing))
```

```python
df=df.dropna()
```

```python
for col in df.columns:
    percent_missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,percent_missing))
```

```python
df['year_correct']=df['year'].astype(str).str[:4]
df
```

```python
df.sort_values(by=['gross'], ascending=False)
```

```python
df.drop_duplicates()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.show()
```

```python
sns.regplot(x=df['budget'],y=df['gross'],data=df)
```

```python
df.corr()
```

```python
cor_mat=df.corr()
sns.heatmap(cor_mat,annot=True)
plt.show()
```

```python
df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized
        
```

```python
cor_mat=df_numerized.corr(method='pearson')
sns.set({'figure.figsize':(15,8)})
sns.heatmap(cor_mat,annot=True, linewidths=.5)

plt.show()
```
